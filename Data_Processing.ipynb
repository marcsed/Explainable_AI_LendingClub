# 2) Data Cleaning

```python
# dropping any "joint" applications - only looking at individual loan applications for this use case
data = data[data['application_type']=='Individual']
data.drop('application_type', axis=1, inplace=True)
```

## Target Variable - "Default"

```python
data['loan_status'].unique()
```

```python
# dropping any active loans (current or late) as I am only considering "matured" loans (defaulted or paid off)
data = data.loc[data['loan_status'].isin(['Fully Paid', 'Charged Off', 'Default'])]

# converting into binary feature
data.replace({'loan_status':{'Fully Paid':0, 'Default':1, 'Charged Off':1}}, inplace=True)

# renaming loan_status to default (target variable)
data.rename(columns = {"loan_status": "default"}, inplace = True)
```

```python
# cleaning categorical columns
data.replace({'term':{' 36 months':0,' 60 months':1}}, inplace=True)

data['grade_code'] = data['grade']
data.replace({'grade':{'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}}, inplace=True)

data['sub_grade_code'] = data['sub_grade']
data.replace({'sub_grade':{'A1':1,'A2':2,'A3':3,'A4':4,'A5':5,
                           'B1':6,'B2':7,'B3':8,'B4':9,'B5':10,
                           'C1':11,'C2':12,'C3':13,'C4':14,'C5':15,
                           'D1':16,'D2':17,'D3':18,'D4':19,'D5':20,
                           'E1':21,'E2':22,'E3':23,'E4':24,'E5':25,
                           'F1':26,'F2':27,'F3':28,'F4':29,'F5':30,
                           'G1':31,'G2':32,'G3':33,'G4':34,'G5':35}}, inplace=True)

data['emp_length_code'] = data['emp_length']
data.replace({'emp_length':{'< 1 year':0.5,'1 year':1,'2 years':2,'3 years':3,'4 years':4,'5 years':5,
                            '6 years':6,'7 years':7,'8 years':8,'9 years':9,'10 years':10,'10+ years':15}}, inplace=True)

data.replace({'verification_status':{'Source Verified':'Source_Verified','Not Verified':'Not_Verified'}}, inplace=True)

data.replace({'initial_list_status':{'f':0,'w':1}}, inplace=True)
```

```python
# clean percentage columns (int_rate, revol_util)
data['int_rate'].replace('%', '', regex=True, inplace=True)
data['revol_util'].replace('%', '', regex=True, inplace=True)
data[['int_rate','revol_util']] = data[['int_rate','revol_util']].apply(pd.to_numeric, errors='coerce', axis=1)
```

# 3) Feature Engineering

### Employment background

```python
# emp_title cleaning
data['emp_title'] = data['emp_title'].str.upper()
data.replace({'emp_title':{'NURSE':'REGISTERED NURSE', 'RN':'REGISTERED NURSE'}}, inplace=True)
data['emp_title'].fillna("MISSING", inplace=True)

# top 15 emp_titles 
data['emp_title'].value_counts()[:15]
titles_top = data['emp_title'].value_counts()[:15].index.tolist()
```

```python
data['manager'] = 0
manager_ind = data[data['emp_title'].str.contains('MANAGER', na=False)].index
data.loc[manager_ind, 'manager'] = 1
data.loc[~data['emp_title'].isin(titles_top), 'emp_title'] = "OTHER"
```

```python
## show example records to verify
data[data['manager']==1]
```

### FICO score

```python
# create new feature "fico" as average of fico_range_high and fico_range_low
data['fico'] = data[['fico_range_high','fico_range_low']].mean(axis=1)
data.drop(['fico_range_high','fico_range_low'], inplace=True, axis=1)
```

### Credit history length

```python
# create new feature "years_cr_line" as years between years of loan issue and earliest credit line
## extract issue year from issue_d
data['issue_d'] = pd.to_numeric(data['issue_d'].str.strip().str[-4:])
## extract year for earliest_cr_line
data['earliest_cr_line'] = pd.to_numeric(data['earliest_cr_line'].str.strip().str[-4:])
## calculate difference and store as yrs_cr_line
data['yrs_cr_line'] = data['issue_d'] - data['earliest_cr_line']
data.drop(['earliest_cr_line'], inplace=True, axis=1)
```

### Loan purpose

```python
# upper case
data['verification_status'] = data['verification_status'].str.upper()
data['purpose'] = data['purpose'].str.upper()
```

# 4) Outlier Handling

```python
### z-score outlier capping (REMOVAL AS ALTERNATIVE)
def zscore_capping(df, cols, thres):
    for col in cols:
        mean = np.mean(df[col])
        std = np.std(df[col])
        
        upper = mean + thres * std
        lower = mean - thres * std
        
        df[col] = np.where(df[col]>upper, upper,
                           np.where(df[col]<lower, lower, df[col]))
        
    return df
```

```python
cont_feats = list(data.select_dtypes('number'))

for col in ['default','term','grade','sub_grade','emp_length','fico']:
    cont_feats.remove(col)
    
cont_feats
```

```python
data = zscore_capping(data, cont_feats, 3)
data.describe()
```

# 5) Missing Data Handling

```python
### DROP ALL ROWS WITH MISSING DATA ### 
# data volume large enough to drop incomplete records as imputation bears risk of adding noise/inaccurate value imputation

data.dropna(inplace=True)
data.info()
```

```python
def get_stats(df, col):
    out = pd.DataFrame(index=['%_missing', 'Mean', 'Median', 'Mode'])
    
    out[col] = [round(df[col].isna().sum()/df.shape[0],4)*100,
                round(df[col].mean(),4),
                round(df[col].median(),4),
                round(df[col].mode()[0],4)]
    
    return out
```
